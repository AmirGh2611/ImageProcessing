## library

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

## reading image

```python
img = cv2.imread("Data/lenna forsen.png")
```

## splitting the channels

```python
z = np.zeros((img.shape[0], img.shape[1]), dtype = "uint8")
B = img[..., 0]
G = img[..., 1]
R = img[..., 2]
```

## Rzz

```python
Rzz = np.stack((R, z, z), axis = 2)
plt.imshow(Rzz)
plt.axis("off");
```

## zGz

```python
zGz = np.stack((z, G, z), axis = 2)
plt.imshow(zGz)
plt.axis("off");
```

## zzB

```python
zzB = np.stack((z, z, B), axis = 2)
plt.imshow(zzB)
plt.axis("off");
```

## RGz

```python
RGz = np.stack((R, G, z), axis = 2)
plt.imshow(RGz)
plt.axis("off");
```

## RzB

```python
RzB = np.stack((R, z, B), axis = 2)
plt.imshow(RzB)
plt.axis("off");
```

## zGB

```python
zGB = np.stack((z, G, B), axis = 2)
plt.imshow(zGB)
plt.axis("off");
```

## RGB

```python
RGB = np.stack((R, G, B), axis = 2)
plt.imshow(RGB)
plt.axis("off");
```
