## library

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## reading image

```python
img = cv2.imread("Data/wallpaper.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

## numper of patches

```python
patch_number_hor = 4
patch_number_ver = 4
```

## patching (no overlap)

```python
height = img.shape[0]
width = img.shape[1]
ls_height = np.linspace(0, height, patch_number_ver, dtype = "int16")
ls_width = np.linspace(0, width, patch_number_hor, dtype = "int16")
```

## result

```python
for i in range(patch_number_hor-1):
    for j in range(patch_number_ver-1):
        patch_img = img[ls_height[i] : ls_height[i+1], ls_width[j] : ls_width[j+1]]
        cv2.imwrite(f"Data/{i}_{j}.jpg", patch_img)
```
