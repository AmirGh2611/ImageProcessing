## Libraries

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Reading Images

```python
img1 = cv2.imread("Data/BrightImage.jpg", 0)
img2 = cv2.imread("Data/DarkImage.jpg", 0)
```

## Q0

```python
plt.figure(figsize = (10, 10))
plt.subplot(121);plt.imshow(img1, cmap = "gray")
plt.subplot(122);plt.imshow(img2, cmap = "gray")
```

## Q1

```python
img1_p_20 = cv2.add(img1, 20)
img1_p_40 = cv2.add(img1, 40)
img1_s_20 = cv2.subtract(img1, 20)
img1_s_40 = cv2.subtract(img1, 40)

img2_p_20 = cv2.add(img2, 20)
img2_p_40 = cv2.add(img2, 40)
img2_s_20 = cv2.subtract(img2, 20)
img2_s_40 = cv2.subtract(img2, 40)

plt.figure(figsize = (20, 7))
plt.subplot(241);plt.imshow(img1_p_20, cmap = "gray");plt.title("img1 + 20")
plt.subplot(242);plt.imshow(img1_p_40, cmap = "gray");plt.title("img1 + 40")
plt.subplot(243);plt.imshow(img1_s_20, cmap = "gray");plt.title("img1 - 20")
plt.subplot(244);plt.imshow(img1_s_40, cmap = "gray");plt.title("img1 - 40")

plt.subplot(245);plt.imshow(img2_p_20, cmap = "gray");plt.title("img2 + 20")
plt.subplot(246);plt.imshow(img2_p_40, cmap = "gray");plt.title("img2 + 40")
plt.subplot(247);plt.imshow(img2_s_20, cmap = "gray");plt.title("img2 - 20")
plt.subplot(248);plt.imshow(img2_s_40, cmap = "gray");plt.title("img2 - 40")
```

## Q2

```python
img1_m_050 = cv2.multiply(img1, 0.5)
img1_m_050 = np.clip(img1_m_050, 0, 255).astype(np.uint8)

img1_m_075 = cv2.multiply(img1, 0.75)
img1_m_075 = np.clip(img1_m_075, 0, 255).astype(np.uint8)

img1_m_125 = cv2.multiply(img1, 1.25)
img1_m_125 = np.clip(img1_m_125, 0, 255).astype(np.uint8)

img1_m_150 = cv2.multiply(img1, 1.5)
img1_m_150 = np.clip(img1_m_150, 0, 255).astype(np.uint8)


img2_m_050 = cv2.multiply(img2, 0.5)
img2_m_050 = np.clip(img2_m_050, 0, 255).astype(np.uint8)

img2_m_075 = cv2.multiply(img2, 0.75)
img2_m_075 = np.clip(img2_m_075, 0, 255).astype(np.uint8)

img2_m_125 = cv2.multiply(img2, 1.25)
img2_m_125 = np.clip(img2_m_125, 0, 255).astype(np.uint8)

img2_m_150 = cv2.multiply(img2, 1.5)
img2_m_150 = np.clip(img2_m_150, 0, 255).astype(np.uint8)

plt.figure(figsize = (20, 7))
plt.subplot(241);plt.imshow(img1_m_050, cmap = "gray");plt.title("img1 * 0.5")
plt.subplot(242);plt.imshow(img1_m_075, cmap = "gray");plt.title("img1 * 0.75")
plt.subplot(243);plt.imshow(img1_m_125, cmap = "gray");plt.title("img1 * 1.25")
plt.subplot(244);plt.imshow(img1_m_150, cmap = "gray");plt.title("img1 * 1.5")

plt.subplot(245);plt.imshow(img2_m_050, cmap = "gray");plt.title("img2 * 0.5")
plt.subplot(246);plt.imshow(img2_m_075, cmap = "gray");plt.title("img2 * 0.75")
plt.subplot(247);plt.imshow(img2_m_125, cmap = "gray");plt.title("img2 * 1.25")
plt.subplot(248);plt.imshow(img2_m_125, cmap = "gray");plt.title("img2 * 1.5")
```

## Q3

```python
img1_a = 255.0 / (img1.max().item()-img1.min().item())
img1_b = -1 * img1_a * img1.min().item()
img1_new = ((img1 * img1_a) + img1_b).astype("uint8")

img2_a = 255.0 / (img2.max().item()-img2.min().item())
img2_b = -1 * img2_a * img2.min().item()
img2_new = ((img2 * img2_a) + img2_b).astype("uint8")

plt.figure(figsize = (30, 10))
plt.subplot(241);plt.imshow(img1, cmap = "gray");plt.title("img1")
plt.subplot(242);plt.imshow(img1_new, cmap = "gray");plt.title("new img1")
plt.subplot(243);plt.imshow(img2, cmap = "gray");plt.title("img2")
plt.subplot(244);plt.imshow(img2_new, cmap = "gray");plt.title("new img2")

plt.subplot(245);plt.hist(img1.ravel(), 256);plt.title("img1 hist")
plt.subplot(246);plt.hist(img1_new.ravel(), 256);plt.title("new img1 hist")
plt.subplot(247);plt.hist(img2.ravel(), 256);plt.title("img2 hist")
plt.subplot(248);plt.hist(img2_new.ravel(), 256);plt.title("new img2 hist")
```

## Q4

```python
img1_08 = np.pow(img1, 0.8)
img1_09 = np.pow(img1, 0.9)
img1_11 = np.pow(img1, 1.1)
img1_12 = np.pow(img1, 1.2)

img2_08 = np.pow(img2, 0.8)
img2_09 = np.pow(img2, 0.9)
img2_11 = np.pow(img2, 1.1)
img2_12 = np.pow(img2, 1.2)

plt.figure(figsize = (20, 7))
plt.subplot(241);plt.imshow(img1_08, cmap = "gray");plt.title("img1 0.8")
plt.subplot(242);plt.imshow(img1_09, cmap = "gray");plt.title("img1 0.9")
plt.subplot(243);plt.imshow(img1_11, cmap = "gray");plt.title("img1 1.1")
plt.subplot(244);plt.imshow(img1_12, cmap = "gray");plt.title("img1 1.2")

plt.subplot(245);plt.imshow(img2_08, cmap = "gray");plt.title("img2 0.8")
plt.subplot(246);plt.imshow(img2_09, cmap = "gray");plt.title("img2 0.9")
plt.subplot(247);plt.imshow(img2_11, cmap = "gray");plt.title("img2 1.1")
plt.subplot(248);plt.imshow(img2_12, cmap = "gray");plt.title("img2 1.2")
```

## Q5

```python
img1_hist_equalized = cv2.equalizeHist(img1)
img2_hist_equalized = cv2.equalizeHist(img2)

plt.figure(figsize = (20, 7))
plt.subplot(241);plt.imshow(img1, cmap = "gray");plt.title("img1")
plt.subplot(242);plt.imshow(img1_hist_equalized, cmap = "gray");plt.title("new img1")
plt.subplot(243);plt.imshow(img2, cmap = "gray");plt.title("img2")
plt.subplot(244);plt.imshow(img2_hist_equalized, cmap = "gray");plt.title("new img2")

plt.subplot(245);plt.hist(img1.ravel(), 256);plt.title("img1 hist")
plt.subplot(246);plt.hist(img1_hist_equalized.ravel(), 256);plt.title("new img1 hist")
plt.subplot(247);plt.hist(img2.ravel(), 256);plt.title("img2 hist")
plt.subplot(248);plt.hist(img2_hist_equalized.ravel(), 256);plt.title("new img2 hist")
```
