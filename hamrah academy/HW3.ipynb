## Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import rank_filter
```

## Reading image

```python
img = cv2.imread("Data/ScannedText.jpg", 0)
```

## Q0

```python
plt.imshow(img, cmap = "gray")
```

## Q1

```python
_, thresh_60 = cv2.threshold(img, 60, 255, cv2.THRESH_BINARY_INV)
_, thresh_120 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY_INV)
_, thresh_180 = cv2.threshold(img, 180, 255, cv2.THRESH_BINARY_INV)

plt.figure(figsize = (20, 7))
plt.subplot(131);plt.imshow(thresh_60, cmap = "gray");plt.title("60 threshold")
plt.subplot(132);plt.imshow(thresh_120, cmap = "gray");plt.title("120 threshold")
plt.subplot(133);plt.imshow(thresh_180, cmap = "gray");plt.title("180 threshold")
```

## Q2

```python
kernel_41 = np.ones((41, 41))
kernel_51 = np.ones((51, 51))
kernel_61 = np.ones((61, 61))

result_41 = cv2.dilate(img, kernel_41, iterations = 1)
result_51 = cv2.dilate(img, kernel_51, iterations = 1)
result_61 = cv2.dilate(img, kernel_61, iterations = 1)

plt.figure(figsize = (20, 7))
plt.subplot(131);plt.imshow(result_41, cmap = "gray");plt.title("41*41 kernel")
plt.subplot(132);plt.imshow(result_51, cmap = "gray");plt.title("51*51 kernel")
plt.subplot(133);plt.imshow(result_61, cmap = "gray");plt.title("61*61 kernel")
```

## Q3

```python
img_s_41 = img - result_41
img_s_51 = img - result_51
img_s_61 = img - result_61

_, thresh_204 = cv2.threshold(img_s_41, 204, 255, cv2.THRESH_BINARY_INV)
_, thresh_202 = cv2.threshold(img_s_51, 202, 255, cv2.THRESH_BINARY_INV)
_, thresh_198 = cv2.threshold(img_s_61, 198, 255, cv2.THRESH_BINARY_INV)

plt.figure(figsize = (20, 20))
plt.subplot(231);plt.imshow(img_s_41, cmap = "gray");plt.title("img-41")
plt.subplot(232);plt.imshow(img_s_51, cmap = "gray");plt.title("img-51")
plt.subplot(233);plt.imshow(img_s_61, cmap = "gray");plt.title("img-61")
plt.subplot(234);plt.imshow(thresh_204, cmap = "gray");plt.title("thresh 204")
plt.subplot(235);plt.imshow(thresh_202, cmap = "gray");plt.title("thresh 202")
plt.subplot(236);plt.imshow(thresh_198, cmap = "gray");plt.title("thresh 198")
```

## Q4

```python
result_1 = rank_filter(img, rank = 7, size = (51, 51))
result_2 = rank_filter(img_s_41, rank = 7, size = (51, 51))
result_3 = rank_filter(img_s_51, rank = 7, size = (51, 51))
result_4 = rank_filter(img_s_61, rank = 7, size = (51, 51))

plt.figure(figsize = (20, 20))
plt.subplot(141);plt.imshow(result_1, cmap = "gray");plt.title("rank original image")
plt.subplot(142);plt.imshow(result_2, cmap = "gray");plt.title("rank img-41")
plt.subplot(143);plt.imshow(result_3, cmap = "gray");plt.title("rank img-51")
plt.subplot(144);plt.imshow(result_4, cmap = "gray");plt.title("rank img-61")
```
